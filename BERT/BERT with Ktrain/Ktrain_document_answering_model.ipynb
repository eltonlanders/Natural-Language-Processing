```python
# Changing directory to our BBC news folder
os.chdir(os.getcwd() + '/bbc')
```

```python
# Initializing the index directory
# Used for indexing all the documents
text.SimpleQA.initialize_index('index') 
```

```python
# Indexing the documents
# The function index_from_folder takes the folder_path where we have all the documents and the index_dir as the parameter
text.SimpleQA.index_from_folder(folder_path='entertainment',index_dir='index')
```

```python
# Creating an instance of the SimpleQA class. Pass the index directory as an argument.
qa = text.SimpleQA('index')
```

```python
# Using the ask functionand asking a question to retrieve answers from the documents for any question
answers = qa.ask('who who had a global hit with where is the love?')
```

```python
# Printing the top 5 answers
qa.display_answers(answers[:5]) # along with the candidate answer, we also get other information such as context, confidence, and document reference.
```

```python
answers = qa.ask('who win at mtv europe awards?')
qa.display_answers(answers[:5])
```
